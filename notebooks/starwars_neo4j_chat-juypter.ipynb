```python
# This Notebook will put Star Wars data into Neo4j, and then Build a Simple ChatBot on top of it.
import os
import pandas as pd
from neo4j import GraphDatabase
from dotenv import load_dotenv
load_dotenv()
```

```python
from yfiles_jupyter_graphs_for_neo4j import Neo4jGraphWidget
from neo4j import GraphDatabase

NEO4J_URI      = os.getenv("NEO4J_URI")
NEO4J_USERNAME = os.getenv("NEO4J_USERNAME")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD")
DATABASE = os.getenv("NEO4J_DATABASE")
driver = GraphDatabase.driver(uri = NEO4J_URI, auth = (NEO4J_USERNAME, NEO4J_PASSWORD), database = DATABASE)

g = Neo4jGraphWidget(driver)

# Give me all the Characters who identify as Human, and return their name, 
cypher_query = """
MATCH (c:Characters)-[:IDENTIFIES_AS]->(s:Species {name: 'Human'})
RETURN c
"""

# Test the query directly
with driver.session() as session:
    result = session.run(cypher_query)
    records = list(result)
    print(f"Query returned {len(records)} results")
    for record in records[:5]:  # Show first 5
        print(record)

widget = g.show_cypher(cypher_query)
display(widget)
```
